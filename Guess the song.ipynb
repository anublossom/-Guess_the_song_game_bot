```python
import sys
import subprocess
import re

# ==========================================
# LINE 1: AUTOMATIC DEPENDENCY CHECKER
# ==========================================
# This safely installs ChromaDB and SentenceTransformers if they are missing
try:
    import chromadb
    from sentence_transformers import SentenceTransformer
    import pandas as pd
except ImportError:
    print("📦 Missing libraries detected. Installing ChromaDB and Transformers for free...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "chromadb", "sentence-transformers", "pandas"])
    import chromadb
    from sentence_transformers import SentenceTransformer
    import pandas as pd

# ==========================================
# LINE 2: LOAD YOUR SCRAPED TAMIL LYRICS FILE
# ==========================================
# Make sure your file name matches exactly what was saved earlier
df = pd.read_csv("rahman_complete_english_tamil.csv")

# ==========================================
# LINE 3: TRANSFORMATION ENGINE FUNCTION
# ==========================================
def clean_lyric_text(text):
    if not isinstance(text, str):
        return ""

    # 1. Strip out 'Male:', 'male:', 'Female:', 'female:' case-insensitively
    text_no_tags = re.sub(r'(?i)(male|female):\s*', '', text)

    # 2. Split into individual lines to keep the song structure intact
    lines = text_no_tags.split('\n')
    cleaned_lines = []

    for line in lines:
        words = line.split()
        word_counts = {}
        line_result = []

        for word in words:
            # Clean punctuation from words to count them accurately
            clean_word = re.sub(r'[^\w]', '', word.lower())
            word_counts[clean_word] = word_counts.get(clean_word, 0) + 1

            # CRUCIAL CONSTRAINT: Drop any word repeating more than 2 times in a row/line
            if word_counts[clean_word] <= 2:
                line_result.append(word)

        cleaned_lines.append(" ".join(line_result))

    return "\n".join(cleaned_lines)

# ==========================================
# LINE 4: PROCESS AND CLEAN THE ENTIRE FILE
# ==========================================
df['Cleaned_Lyrics'] = df['Lyrics'].apply(clean_lyric_text)

# ==========================================
# LINE 5: INITIALIZE THE LOCAL CHROMADB STORE
# ==========================================
# This sets up a permanent database folder named 'rahman_vector_db'
chroma_client = chromadb.PersistentClient(path="rahman_vector_db")

# ==========================================
# LINE 6: CREATE A CLEAN CORE REGISTRY POOL
# ==========================================
collection = chroma_client.get_or_create_collection(name="rahman_tamil_songs")

# ==========================================
# LINE 7: INITIALIZE THE EMBEDDING BRAIN
# ==========================================
# This pulls down a multilingual model to read and translate your English-script Tamil text
embedding_model = SentenceTransformer("intfloat/multilingual-e5-base")

# ==========================================
# LINE 8: CONVERT FULL LYRICS AND POPULATE DB
# ==========================================
print("\n🔄 Generating vector embeddings. Please wait...")

for index, row in df.iterrows():
    lyric_content = row['Cleaned_Lyrics']

    # Skip processing if a row contains empty lyric values
    if not lyric_content.strip():
        continue

    # Translate text layout into mathematical matrix arrays
    vector_embedding = embedding_model.encode(lyric_content).tolist()

    # Map out metadata using the other 3 columns for game reference
    metadata_package = {
        "title": row['Song Title'],
        "movie": row['Movie/Album'],
        "composer": row['Composer']
    }

    # Securely append inside ChromaDB
    collection.add(
        ids=[f"song_{index}"],
        embeddings=[vector_embedding],
        documents=[lyric_content], # The entire document goes here for robust context search
        metadatas=[metadata_package]
    )

print(f"\n🎉 Complete! Your cleaned database has been embedded successfully.")
print(f"📦 ChromaDB total records stored: {collection.count()} songs.")
```

```python
# ==========================================
# LINE 1: INSTALL LANGGRAPH & GROQ DRIVERS
# ==========================================
!pip install -q langgraph langchain-groq

from typing import TypedDict, List
import random
import re
import chromadb
from sentence_transformers import SentenceTransformer
from langgraph.graph import StateGraph, START, END
from langchain_groq import ChatGroq

# ==========================================
# LINE 2: INSERT YOUR FREE GROQ KEY HERE
# ==========================================
MY_FREE_KEY = "YOUR_API_KEY"

# 1. Connect back to your existing 154-song ChromaDB folder
chroma_client = chromadb.PersistentClient(path="rahman_vector_db")
collection = chroma_client.get_collection(name="rahman_tamil_songs")
embedding_model = SentenceTransformer("intfloat/multilingual-e5-base")

# 2. Connect to the high-speed free Cloud Llama-3.1 model brain
llm = ChatGroq(model="llama-3.1-8b-instant", groq_api_key=MY_FREE_KEY, temperature=0.4)

class GameState(TypedDict):
    raw_four_words: str      # Step 1: The exact 4-word phrase pulled from the song
    ai_clue: str             # Step 2: The intelligent clue created by Agent 2
    played_songs: List[str]
    correct_answer: str
    movie_name: str
    user_guess: str
    attempts: int
    score: int

# ==========================================
# STEP 1: THE DATABASE NODE (Pure Song Text)
# ==========================================
def fetch_exact_song_phrase_node(state: GameState):
    """Pulls a song and extracts an exact, contiguous 4-word line from the lyrics."""
    current_lyrics = state.get("raw_four_words", "love romance rain")
    played_history = state.get("played_songs", [])

    query_vector = embedding_model.encode(current_lyrics).tolist()
    results = collection.query(query_embeddings=[query_vector], n_results=5)

    selected_doc = ""
    selected_title = ""
    selected_movie = ""

    # FIX: ChromaDB queries return arrays inside an extra outer list, we unpack the first index [0] safely
    if results and 'documents' in results and len(results['documents']) > 0 and len(results['documents'][0]) > 0:
        docs_list = results['documents'][0]
        metadatas_list = results['metadatas'][0]

        for i in range(len(docs_list)):
            title = metadatas_list[i]['title']
            if title not in played_history:
                selected_doc = docs_list[i]
                selected_title = title
                selected_movie = metadatas_list[i]['movie']
                break

    # Safety fallback if every matching option was already used or if database parsing hit a glitch
    if not selected_doc:
        all_records = collection.get()
        if all_records and len(all_records['documents']) > 0:
            random_index = random.randint(0, len(all_records['documents'])-1)
            selected_doc = all_records['documents'][random_index]
            selected_title = all_records['metadatas'][random_index]['title']
            selected_movie = all_records['metadatas'][random_index]['movie']
        else:
            # Absolute baseline backup if database can't find anything
            selected_doc = "chinna chinna aasai siragadikki aasai\nmuthu muthu aasai"
            selected_title = "Chinna Chinna Aasai"
            selected_movie = "Roja"

    # Extract 4 contiguous words exactly as written in a song line
    lines = [line.strip() for line in selected_doc.split('\n') if len(line.strip().split()) >= 4]
    if lines:
        chosen_line = random.choice(lines)
        words = chosen_line.split()
        exact_phrase = " ".join(words[:4])
    else:
        # Fallback if text layout has short lines
        words = selected_doc.split()
        exact_phrase = " ".join(words[:4]) if len(words) >= 4 else " ".join(words)

    return {
        "raw_four_words": exact_phrase,
        "correct_answer": selected_title,
        "movie_name": selected_movie,
        "played_songs": played_history + [selected_title],
        "attempts": 0,
        "score": state.get("score", 0)
    }

# ==========================================
# STEP 2: AGENT 2 NODE (The Clue Builder)
# ==========================================
def generate_ai_clue_node(state: GameState):
    """Agent 2 reads the real 4-word lyric line and turns it into an engaging clue."""
    phrase = state["raw_four_words"]

    prompt = f"""
    You are an expert Tamil music game host.
    Here is an exact 4-word phrase taken directly from an A.R. Rahman song line:
    "{phrase}"

    Task: Write an engaging, helpful trivia clue for the player based on this phrase.
    Rules:
    - You can translate the phrase to English, describe its mood, or hint at the style.
    - Do NOT mention the song name '{state['correct_answer']}' or movie name '{state['movie_name']}'.
    - Keep your clue concise (1 to 2 short sentences max).

    Your Trivia Clue:"""

    ai_response = llm.invoke(prompt)
    clue_text = ai_response.content.strip()

    print(f"\n🎵 Real Lyric Line: \"{phrase}\"")
    print(f"💡 Agent 2 Clue: {clue_text}")
    return {"ai_clue": clue_text}

def wait_for_user_guess_node(state: GameState):
    guess = input("⌨️ Enter your guess (Song Title or Movie Name): ")
    return {"user_guess": guess}

# ==========================================
# STEP 3: AGENT 3 NODE (The Evaluator Edge)
# ==========================================
def evaluate_guess_edge(state: GameState):
    guess = state["user_guess"].strip().lower()
    answer_title = state["correct_answer"].strip().lower()
    answer_movie = state["movie_name"].strip().lower()
    current_attempts = state["attempts"] + 1
    current_score = state.get("score", 0)

    if guess == answer_title or guess == answer_movie:
        print(f"🎉 Correct! Agent 3 confirms it was '{state['correct_answer']}' from '{state['movie_name']}'.")
        state["score"] = current_score + 1
        return "next_round"
    elif current_attempts >= 3:
        print(f"❌ Game Over! Evaluator reveals: The answer was '{state['correct_answer']}' from '{state['movie_name']}'.")
        print(f"🏆 Final Score: {current_score} points.")
        return "end_game"
    else:
        print(f"❌ Wrong! Evaluator rejects guess. Try again. (Attempts: {current_attempts}/3)")
        state["attempts"] = current_attempts
        return "retry_guess"

# ==========================================
# BUILDING THE COMPILING MAP
# ==========================================
workflow = StateGraph(GameState)

workflow.add_node("fetch_phrase", fetch_exact_song_phrase_node)
workflow.add_node("make_clue", generate_ai_clue_node)
workflow.add_node("get_input", wait_for_user_guess_node)

workflow.add_edge(START, "fetch_phrase")
workflow.add_edge("fetch_phrase", "make_clue")
workflow.add_edge("make_clue", "get_input")

workflow.add_conditional_edges(
    "get_input",
    evaluate_guess_edge,
    {
        "next_round": "fetch_phrase",
        "retry_guess": "get_input",
        "end_game": END
    }
)

game_app = workflow.compile()
print("⚙️ New 3-Layer Meaningful Game Engine Compiled Successfully!")
```
